```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv(r'./datainput/time_series_covid19_confirmed_global.csv', encoding = "ISO-8859-1")
df.head()
```

```python
total_cols = df.keys()
print(len(total_cols))
print(len(df))
```

```python
prov = []
country = []
lat = []
lon = []
date=[]
val = []
```

```python
%%time
fila = 0
for idx,item in df.iterrows():
    fila += 1
    for coldate in total_cols[4:]:
        prov.append(item['Province/State'])
        if str(carga.loc[100].provincia) == 'nan':
            country.append(item['Country/Region'])
        else:
            country.append(item['Country/Region'] + '(' + item['Province/State'] + ')')
        lat.append(item['Lat'])
        lon.append(item['Long'])
        date_time_obj = datetime. strptime(coldate, '%m/%d/%y')
        date.append(date_time_obj)
        val.append(item[coldate])
        
    
    #print(item['Lat'])
    #break
    
    
fila
```

```python
len(prov),len(country),len(lat),len(lon),len(date),len(val)
```

```python
len(date)
carga = pd.DataFrame({})
d = {'provincia':prov, 'country': country, 'lat': lat, 'long': lon, 'dates': date, 'value':val}
carga = pd.DataFrame(data=d)
#for coldate in total_cols[4:]:
#    print(coldate)
```

```python
carga
```

```python
resumen = carga.groupby(['dates']).sum()
resumen = resumen.reset_index()
resumen['provincia'] = ''
resumen['country'] = '-Global-'
resumen['lat'] = 0
resumen['long'] = 0
resumen
```

```python
carga.append(resumen)
```
